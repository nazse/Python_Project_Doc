```python
# Change string variables to binary categorical (0, 1) for model fitting
#from sklearn.preprocessing import LabelEncoder 
lb_make = LabelEncoder()
my_data_frame['Amyloid_Pathology'] = lb_make.fit_transform(my_data_frame['Amyloid_Pathology'])
my_data_frame['T_Tau_Pathology'] = lb_make.fit_transform(my_data_frame['T_Tau_Pathology'])
my_data_frame['P_Tau_Pathology'] = lb_make.fit_transform(my_data_frame['P_Tau_Pathology'])
my_data_frame['Apoe4'] = lb_make.fit_transform(my_data_frame['Apoe4'])
my_data_frame['SEX'] = lb_make.fit_transform(my_data_frame['SEX'])
my_data_frame['livealone'] = lb_make.fit_transform(my_data_frame['livealone'])
my_data_frame['DEMIS'] = lb_make.fit_transform(my_data_frame['DEMIS'])
my_data_frame['ANYDEPK'] = lb_make.fit_transform(my_data_frame['ANYDEPK'])

# Format our dependent variables for a "For Loop and If esle statements"
response_var = ["Amyloid_Pathology","T_Tau_Pathology","P_Tau_Pathology"]
CSF_target = my_data[response_var]
```

```python
for target in CSF_target:
    if target == "Amyloid_Pathology":
        print('Program is running Model 1')
        print('')
        model= smf.logit(formula = target + "~SEX + AGEINT + Apoe4 + DBP + SBP +MIN66 + ANYDEPK + MIN17 + MIN66 + RISK6 + MRI391 + MRI392 + MRI393 + MRI394 + MRI395 + MRI396", data= my_data_frame).fit()
        # GETTING THE ODDS RATIOS, AND 95% CI
        model_odds = pd.DataFrame(np.exp(model.params), columns= ['OR'])
        model_odds[['2.5%', '97.5%']] = np.exp(model.conf_int())
        print('')
        print('Model 1: CSF %s Modeling in Relation to Basal Forebrain and Demorgraphic Variables ....'%(target))
        print('')
        print(model.summary())
        print('')
        print('Odd Ratio (OR) and  (CI; 2.5%, 97.5%) for the CSF Amyloid Pathology Model 1 ...')
        print('')
        print(model_odds)
        print('')
    elif target == "T_Tau_Pathology":
        print('Program is running Model 2')
        print('')
        model= smf.logit(formula = target + "~SEX + AGEINT + Apoe4 + DBP + SBP +MIN66 + ANYDEPK + MIN17 + MIN66 + RISK6 + MRI391 + MRI392 + MRI393 + MRI394 + MRI395 + MRI396", data= my_data_frame).fit()
        # GETTING THE ODDS RATIOS, AND 95% CI
        model_odds = pd.DataFrame(np.exp(model.params), columns= ['OR'])
        model_odds[['2.5%', '97.5%']] = np.exp(model.conf_int())
        print('')
        print('Model 2: CSF %s Modeling in Relation to Basal Forebrain and Demorgraphic Variables ....'%(target))
        print('')
        print(model.summary())
        print('')
        print('Odd Ratio (OR) and  (CI; 2.5%, 97.5%) for the CSF Total Tau Pathology Model 2 ...')
        print('')
        print(model_odds)
    elif target == "P_Tau_Pathology":
        print('Program is running Model 3')
        print('')
        model= smf.logit(formula = target + "~SEX + AGEINT+ Apoe4 + DBP + SBP + MIN66 + MIN17 + RISK6 + MRI391 + MRI392 + MRI393 + MRI394 + MRI395 + MRI396", data= my_data_frame).fit()
        # GETTING THE ODDS RATIOS, AND 95% CI
        model_odds = pd.DataFrame(np.exp(model.params), columns= ['OR'])
        model_odds[['2.5%', '97.5%']] = np.exp(model.conf_int())
        print('')
        print('Model 3: CSF %s Modeling in Relation to Basal Forebrain and Demorgraphic Variables ....'%(target))
        print('')
        print(model.summary())
        print('')
        print('Odd Ratio (OR) and  (CI; 2.5%, 97.5%) for the CSF Phospho Tau Pathology Model 3 ...')
        print('')
        print(model_odds)
    else:
        pass
    
    
```
